```python
from utils import create_new_csv
from grid import * 
asc_df = pd.read_csv("/home/uwm/EGMS/dane_pociete/ascLodz.csv", delimiter=';', decimal=',')
dsc_df = pd.read_csv("/home/uwm/EGMS/dane_pociete/desc_Lodz.csv", delimiter=';', decimal=',')
asc_cols = get_date_columns(asc_df.columns)
create_new_csv("OutputGrid_Lodz.csv", asc_cols, ';')
print(len(asc_df), len(dsc_df))
data = grid(asc_df, dsc_df , 0.05)

```

```python
import matplotlib.pyplot as plt
x = [point[0] for point in data]
y = [point[1] for point in data]
values = [point[2] for point in data]

# Create a scatter plot
plt.scatter(x, y, s=values)  # 's' determines the size of each point based on 'value'

# Add labels and title
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.title('2D Scatter Plot')

# Show the plot
plt.show()
```
