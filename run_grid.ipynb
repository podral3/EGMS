```python
from grid import * 
asc_df = pd.read_csv("/home/EGMS/data/102_100.csv", nrows=10000)
dsc_df = pd.read_csv("/home/EGMS/data/153_100.csv", nrows=10000)
print(len(asc_df), len(dsc_df))
data = grid(asc_df, dsc_df , 0.5)

```

```python
import matplotlib.pyplot as plt
x = [point[0] for point in data]
y = [point[1] for point in data]
values = [point[2] for point in data]

# Create a scatter plot
plt.scatter(x, y, s=values)  # 's' determines the size of each point based on 'value'

# Add labels and title
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.title('2D Scatter Plot')

# Show the plot
plt.show()
```
